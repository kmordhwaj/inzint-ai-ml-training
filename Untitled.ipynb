```python
# A -> rainfall more than 500mm in june
# B -> flooded

# bayes theorem
# p(A/B) = (p(B/A). p(A))/(p(B/A). p(A) + p(B/!A).p(!A))
# p(B/A) = p(A and B)/p(A)

pA = round(93/118, 3)
pAnot = 1 - pA
# p(B/A) = p(A and B)/p(A)
pBgivenA = round(54/118, 3)/pA
# p(B/!A) = p(!A and B)/p(!A)
pBgivenAnot = round(6/118, 3)/(pAnot)

pAgivenB = (pBgivenA * pA)/(pBgivenA * pA + pBgivenAnot * pAnot)
pAgivenB
```

```python
# A -> rainfall more than 500mm in june
# B -> flooded

# bayes theorem
# p(B/A) = (p(A/B). p(B))/(p(A/B). p(B) + p(A/!B).p(!B))
# p(A/B) = p(A and B)/p(B)

pB = round(60/118, 3)
pBnot = 1 - pB
# p(A/B) = p(A and B)/p(B)
pAgivenB = round(54/118, 3)/pB
# p(A/!B) = p(A and !B)/p(!B)
pAgivenBnot = round(39/118, 3)/(pBnot)

pBgivenA = (pAgivenB * pB)/(pAgivenB * pB + pAgivenBnot * pBnot)
pBgivenA
```

```python
# conclusion
# probability of flood, given rainfall in june is greater than 500mm is 0.58
```
