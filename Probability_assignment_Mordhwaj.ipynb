```python
# Load the “kerala.csv”data into a DataFrame and perform the following tasks:

# 1.Explore the DataFrame using info()and describe()functions

# 2.June and July are the peak months of rainfall. Consider that if it rains more than 500mm, then the chances of flood become more; create a Datarame with columns –“YEAR”, “JUN_GT_500” (Contains a boolean value to show whether it rained more than 500 mm in June), “JUL_GT_500” (Contains a boolean value to show whether it rained more than 500 mm in July),and “FLOODS” (Contains a boolean value to show whether it flooded that year)

# 3.Calculate the probability of flood given it rained more than 500 mm in June (P(A|B))

# 4.Calculate the probability of rain more than 500 mm in June, given it flooded that year (P(B|A))

# 5.Probability of flood given it rained more than 500 mm in July6.Probability of rain of more than 500 mm in July given it flooded that year (P(B|A))
```

```python
import pandas as pd

df=pd.read_csv("kerala_floods.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
jun = (df["JUN"] > 500).astype('int')
jun
```

```python
july = (df["JUL"] > 500).astype('int')
july
```

```python
flooded = df["FLOODS"].map({"YES":1, "NO": 0})
flooded
```

```python
df["JUN_GT_500"] = jun
df["JUL_GT_500"] = july
df["FLOODS"] = flooded
df
```

```python
data = df.loc[:, ["YEAR", "JUN_GT_500","JUL_GT_500","FLOODS" ]]
data
```

```python
pd.crosstab(data["FLOODS"], data["JUN_GT_500"], margins=True, margins_name="Total")
```

```python
# A -> rainfall more than 500mm in june
# B -> flooded

# bayes theorem
# p(A/B) = (p(B/A). p(A))/(p(B/A). p(A) + p(B/!A).p(!A))
# p(B/A) = p(A and B)/p(A)

pA = round(93/118, 3)
pAnot = 1 - pA
# p(B/A) = p(A and B)/p(A)
pBgivenA = round(54/118, 3)/pA
# p(B/!A) = p(!A and B)/p(!A)
pBgivenAnot = round(6/118, 3)/(pAnot)

pAgivenB = (pBgivenA * pA)/(pBgivenA * pA + pBgivenAnot * pAnot)
pAgivenB
```

```python
# consclusion
# prob of rainfall, more than 500mm on june given that year flooded is 0.8998
```

```python
# A -> rainfall more than 500mm in june
# B -> flooded

# bayes theorem
# p(B/A) = (p(A/B). p(B))/(p(A/B). p(B) + p(A/!B).p(!B))
# p(A/B) = p(A and B)/p(B)

pB = round(60/118, 3)
pBnot = 1 - pB
# p(A/B) = p(A and B)/p(B)
pAgivenB = round(54/118, 3)/pB
# p(A/!B) = p(A and !B)/p(!B)
pAgivenBnot = round(39/118, 3)/(pBnot)

pBgivenA = (pAgivenB * pB)/(pAgivenB * pB + pAgivenBnot * pBnot)
pBgivenA
```

```python
# consclusion
# prob of floods, given in that year june had more than 500mm of rain is 0.58
```
