# Problem 1 :  Zscore

```python
# Zscore
# Weight of the student is 60, find out the how many standard deviation away the point is located from mean where mean is 72.9 and standard deviation is 8.63 as per the weights data of students
# Considering the weight data of student with mean 72.9 and standard deviation of 8.63, what is the probability of selecting a student having weight less than 65?
```

```python
import scipy.stats as stats

mean = 72.9
std_dev = 8.63
weight = 65

# z-score for weight 65
z_score = (weight - mean) / std_dev

# probability using the cumulative distribution function
probability = stats.norm.cdf(z_score)
probability
```

# Problem 2 : Z distibution

```python
# Z distribution
# Load the exam.csv dataset into a DataFrame and perform the following tasks on the “math score,” ”readingscore,” and “writing score” columns:

# Plot the distributions using the seabornlibrary

# Find the mean and standard deviation using the scipylibrary

# Find skewness and kurtosisusing the stats module from the scipylibrary

# Find Standard Normal Distribution(z-score) using the stats module from the scipyl ibrary

# Plot the Z distribution using the seaborn library
```

```python
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt

df = pd.read_csv("exam.csv")
df
```

```python
# Plot the distributions using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df["math score"], kde=True, color="blue")
plt.title("Math Score Distribution")

plt.subplot(1, 3, 2)
sns.histplot(df["reading score"], kde=True, color="red")
plt.title("Reading Score Distribution")

plt.subplot(1, 3, 3)
sns.histplot(df["writing score"], kde=True, color="green")
plt.title("Writing Score Distribution")

plt.tight_layout()
plt.show()
```

```python
# mean and standard deviation
math_mean = df["math score"].mean()
math_std_dev = df["math score"].std()
reading_mean = df["reading score"].mean()
reading_std_dev = df["reading score"].std()
writing_mean = df["writing score"].mean()
writing_std_dev = df["writing score"].std()

math_mean
```

```python
# skewness and kurtosis
math_skewness = stats.skew(df["math score"])
math_kurtosis = stats.kurtosis(df["math score"])
reading_skewness = stats.skew(df["reading score"])
reading_kurtosis = stats.kurtosis(df["reading score"])
writing_skewness = stats.skew(df["writing score"])
writing_kurtosis = stats.kurtosis(df["writing score"])

math_skewness
```

```python
# standard Normal Distribution (z-score)
math_z_scores = stats.zscore(df["math score"])
reading_z_scores = stats.zscore(df["reading score"])
writing_z_scores = stats.zscore(df["writing score"])

math_z_scores
```

```python
# Plot Z distributions
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(math_z_scores, kde=True, color="blue")
plt.title("Math Score Z Distribution")

plt.subplot(1, 3, 2)
sns.histplot(reading_z_scores, kde=True, color="red")
plt.title("Reading Score Z Distribution")

plt.subplot(1, 3, 3)
sns.histplot(writing_z_scores, kde=True, color="green")
plt.title("Writing Score Z Distribution")

plt.tight_layout()
plt.show()
```
