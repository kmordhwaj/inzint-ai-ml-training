```python
students_perf = pd.read_csv('StudentsPerformance.csv')
```

```python
students_perf
```

```python
val = students_perf.groupby('race/ethnicity')['math score'].mean()
```

```python
x = val.index
y = val.values
plt.bar(x, y)
plt.show()
```

```python
languages = ["c++", "C", "python", "java"]
popularity= [22.2, 43.5, 34.5, 6.7]
plt.figure(figsize=(5,5))
plt.pie(popularity, labels=languages, explode=[0, 0.2, 0, 0], radius=1, autopct='%1.1%%')
plt.show()
```
